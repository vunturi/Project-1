```python
# Creating DataFrame
restaurant_list_df = pd.DataFrame(restaurant_list)
restaurant_list_df.head(2)
```

```python
# Repositioning headers without using .drop method on columns
organized_rest_df = restaurant_list_df[["name", "display_phone", "location", "price", "rating", "review_count"]]
organized_rest_df.head()
```

```python
# Extract only Zip Code instead of full address under location column
for x in range(0,organized_rest_df.shape[0]):
    organized_rest_df.loc[x, 'location'] = restaurant_list[x]['location']['zip_code'] 
```

```python
# Rename "location", "display_phone", "review_count" to zip_code, phone and reviews
organized_rest_df = organized_rest_df.rename(index = str, columns = {"location": "zip_code", "display_phone": "phone", "review_count": "reviews"}) 
organized_rest_df.head()
```
