```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json

# Google developer API key
from gconfig import gkey
```

```python
# geocoordinates
target_coordinates = "32.7766642,-96.79698789999999"
target_radius = 50
target_type = "restaurant"

# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "radius": target_radius,
    "type": target_type,
    "key": gkey
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
response = requests.get(base_url, params=params)
```

```python
# convert response to json
places = response.json()

# Print the json (pretty printed)
len(json.dumps(places, sort_keys=True))
```

```python
print(json.dumps(places, indent=10, sort_keys=True))
```

```python
restaurants = []

for restaurant in places['name']: 
    rest_name = places['name']
    restaurants.append(rest_name)
    
print(restaurants)
    
```
