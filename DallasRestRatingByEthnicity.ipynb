```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mtick
```

```python
rating_file = pd.read_csv("ratingbyethnicity.csv", encoding="utf-8")
```

```python
df = pd.DataFrame(rating_file)
df
```

```python
df.plot(kind='bar',x='ethnicity',y='ave_rating', facecolor = "lightgreen")
plt.rcParams["figure.figsize"] = (25,10)
plt.xticks(rotation = 45, ha='right', fontsize = 20)
plt.title("Average Rating By Ethnicity",fontsize = 34, fontweight = 'bold')
plt.ylabel("Rating",fontsize = 25)
plt.xlabel("Ethnicity", fontsize = 25)
plt.legend(fontsize = 12)
plt.show()
```
